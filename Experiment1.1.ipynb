```python
# Make a list of all the possible decks, the last element is used when no cabin code is present...
cabin_list = ['A', 'B', 'C', 'D', 'E', 'F', 'T', 'G', 'Unknown']
```

```python
# Define a function to search for a substring in a list
def search_substring(big_string, substring_list):
    for substring in substring_list:
        if substring in big_string:
            return substring
    return substring_list[-1]
```

```python
# replace passenger's name with his/her title (Mr, Mrs...)
def get_title(string):  
    import re
    
# Declared a function ("get_title" is a function, it has a p[arameter string and that contains passenger's name...)
    
    regex = re.compile(r'Mr|Don|Major|Capt|Jonkheer|Rev|Col|Dr|Mrs|Countess|Dona|Mme|Ms|Miss|Mlle|Master', re.IGNORECASE)
    # 
    results = regex.search(string)
    if results != None:
        return (results.group().lower())
    else:
        return(str(np.nan))
```

```python
# Test the get_title function
get_title("Wilkes, Mrs. James (Ellen Needs)	")
```

```python
# dictionary to map to generate the new feature vector
title_dictionary = {
"capt":"Officer",
"col":"Officer",
"major":"Officer",
"dr":"Officer",
"jonkheer":"Royalty",
"rev":"Officer",
"countess":"Royalty",
"dona":"Royalty",
"lady":"Royalty",
"don":"Royalty",
"mr":"Mr",
"mme":"Mrs",
"ms":"Mrs",
"mrs":"Mrs",
"miss":"Miss",
"mlle":"Miss",
"master":"Master",
"nan":"Mr"
}
```

```python
# Extract Deck information from Cabin column
test['Deck'] = test['Cabin'].map(lambda x : search_substring(str(x), cabin_list))
```

```python
# Drop the original Cabin column
test.drop('Cabin', axis = 1, inplace = True)
```

```python
# Display the head of the DataFrame after dropping 'Cabin'
test.head()
```

```python
# Apply the get_title function to create a new 'Title' column
test['Title'] = test['Name'].apply(get_title)
```

```python
# Map titles to generalized categories using the dictionary
test['Title'] = test['Title'].map(title_dictionary)
```

```python
# Display the head of the DataFrame after 'Title' mapping
test.head()
```

```python
title_list = ['Mr', 'Mrs', 'Master', 'Royalty', 'Officer']

def age_nan_replace(means, dframe, title_list):
    # Iterate through each title in the title_list
    for title in title_list:  # Added this loop to define 'title' variable
        temp = dframe['Title'] == title
        dframe.loc[temp, 'Age'] = dframe.loc[temp, 'Age'].fillna(means[title])
        
age_nan_replace(means_title, test, title_list)
# age_nan_replace(means_title, title_list)
```
